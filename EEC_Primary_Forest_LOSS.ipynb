## การสูญเสียพื้นที่ป่าปฐมภูมิ (Primary forest) และการปล่อย CO2e ภายในขอบเขตพื้นที่ EEC ต้ังแต่ พ.ศ. 2544 - 2567 ข้อมูลจาก Global Forest Watch (GFW) University of Maryland and World Resources Institute. "Global Primary Forest Loss". Accessed through Global Forest Watch on 30/05/2025 from www.globalforestwatch.org. เข้าถึงได้ที่ https://gfw.global/4kKkzNH วิเคราะห์เบื้องต้นใน MS-Excel แล้วดังนี้ https://docs.google.com/spreadsheets/d/1vO9Vhvn0FwwLIQcztwSq9fTFZH8exBhF/edit?usp=drive_link&ouid=109409066129110868195&rtpof=true&sd=true

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# ข้อมูล
data = {
    'area_ha': [
        9.443, 27.359, 145.399, 79.416, 93.696, 165.062,
        138.091, 49.695, 92.802, 105.964, 102.990, 142.261,
        27.825, 14.245, 4.801, 28.492, 5.774, 12.828,
        39.018, 29.862, 14.763, 12.744, 22.563, 50.481
    ],
    'co2e_mg': [
        4919.83, 14245.39, 79567.36, 41157.53, 53682.97, 93225.15,
        75186.04, 27075.87, 50549.04, 59346.21, 59478.78, 81773.19,
        15926.03, 7619.72, 2654.28, 16117.74, 3433.56, 8059.77,
        21583.58, 18415.51, 8559.87, 9154.31, 15279.90, 30185.51
    ]
}

df = pd.DataFrame(data)

# สร้างโมเดล regression
X = df[['area_ha']]
y = df['co2e_mg']
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)

# พารามิเตอร์สมการ
slope = model.coef_[0]
intercept = model.intercept_
r2 = model.score(X, y)

# สร้างกราฟ
plt.figure(figsize=(10, 6))
plt.scatter(df['area_ha'], df['co2e_mg'], color='green', label='Data Points')
plt.plot(df['area_ha'], y_pred, color='red', label='Regression Line')
plt.xlabel('Area of Primary Forest Loss (ha)')
plt.ylabel('Gross Emission of CO2e (Mg)')
plt.title('Linear Regression of PRIMARY FOREST LOSS and CO2e Emission\nfrom EEC Area Thailand 2001 - 2024')
equation = f'y = {slope:.2f}x + {intercept:.2f}\n$R^2$ = {r2:.3f}'
plt.text(10, max(y)*0.8, equation, fontsize=12, bbox=dict(facecolor='white'))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

จุดสีเขียวคือข้อมูลจริงของพื้นที่ป่าที่สูญเสียกับการปล่อย CO2e

จากผลการวิเคราะห์ Linear Regression ระหว่างการสูญเสียพื้นที่ป่าปฐมภูมิ (ha) กับ การปล่อยก๊าซเรือนกระจก (CO₂e, Mg) ในพื้นที่ EEC ของประเทศไทย ระหว่างปี พ.ศ. 2544–2567 (2001–2024) สามารถอภิปรายได้ดังนี้:
สมการถดถอยเชิงเส้น (Linear Regression Equation)
y = 568.35x + 1253.17
y: ปริมาณการปล่อย CO₂e (ล้านกรัม หรือ Mg)
x: พื้นที่ป่าปฐมภูมิที่สูญเสียไป (เฮกตาร์)
R² = 0.975

ความหมายของสมการ
Slope (ค่าความชัน) ≈ 568.35
หมายถึง ทุกๆ การสูญเสียพื้นที่ป่าปฐมภูมิ 1 ha จะส่งผลให้เกิดการปล่อย CO₂e เพิ่มขึ้นประมาณ 568.35 Mg

Intercept (จุดตัดแกน Y) ≈ 1253.17
หมายถึง หากไม่มีการสูญเสียพื้นที่ป่าเลย (x = 0) จะยังคงมีการปล่อย CO₂e อยู่ที่ประมาณ 1,253 Mg อันอาจเกิดจากแหล่งอื่น (หรือ baseline emission ที่ไม่เกี่ยวกับการสูญเสียป่าโดยตรง)

ค่า R² = 0.975
หมายถึง 97.5% ของความแปรปรวนในการปล่อย CO₂e สามารถอธิบายได้ด้วยการสูญเสียพื้นที่ป่าปฐมภูมิ
→ เป็นความสัมพันธ์เชิงเส้นที่ "เข้มแข็งมาก" แสดงว่า loss ของ primary forest มีอิทธิพลสูงต่อการปล่อย CO₂e

ข้อสังเกตจากข้อมูล
ในช่วงปีที่มีการสูญเสียป่าเป็นจำนวนมาก เช่น ปี พ.ศ. 2546, 2549, 2555 จะพบว่ามีการปล่อย CO₂e สูงมากเช่นกัน ซึ่งสนับสนุนความสัมพันธ์เชิงบวกที่ได้จากการวิเคราะห์

ช่วงปีที่การสูญเสียลดลง (พ.ศ. 558 - 2561) ก็มีการปล่อย CO₂e ลดลงตามลำดับ

ข้อสรุป
การวิเคราะห์นี้ ยืนยัน ว่าการสูญเสียพื้นที่ป่าปฐมภูมิในเขต EEC มีความสัมพันธ์อย่างชัดเจนกับการปล่อย CO₂e

การลดการตัดไม้ทำลายป่าจะช่วยลดการปล่อย CO₂e ได้โดยตรง ซึ่งสอดคล้องกับเป้าหมายการพัฒนาที่ยั่งยืน (SDG 13: Climate Action)




อภิปรายในเชิงนโยบาย โดยอิงตาม ตัวชี้วัดของแผนพัฒนาสิ่งแวดล้อม EEC ซึ่งระบุว่า:

“พื้นที่ป่าไม้ (ทั้งป่าอนุรักษ์ ป่าเศรษฐกิจ ป่าชายเลน และพื้นที่สีเขียวในเมือง) ต้องเพิ่มขึ้นอย่างน้อย 5% ต่อปี” ที่มา; โครงการจัดทำแผนสิ่งแวดล้อมในพื้นที่เขตพัฒนาเศรษฐกิจพิเศษภาคตะวันออก ระยะที่ 2 (พ.ศ. 2565 - 2569) โดยสำนักงานนโยบายและแผนทรัพยากรธรรมชาติและสิ่งแวดล้อม (สผ.)https://www.onep.go.th/wp-content/uploads/2023/05/performance-eec-audit65.pdf?fbclid=IwZXh0bgNhZW0CMTAAYnJpZBExM0FrUUdJQWlNZWxXZ09UUgEeR_o3xIQRwVeUWo_BIew1dgGmIsLhxd9HT5KSP9eK67rKN2OKuNS6a7YaJgE_aem_T3QBDvLL2zk6SHVVSxac3g

ข้อสังเกต:
แม้จะไม่มีข้อมูลพื้นที่ป่าที่เพิ่มขึ้นโดยตรง แต่จากตัวเลข "พื้นที่ป่าที่สูญเสีย" ยังเพิ่มขึ้นในบางปี ซึ่ง ขัดกับเป้าหมาย ที่พื้นที่ป่าควร “เพิ่มขึ้น” 5% ต่อปี

การสูญเสียเพิ่มขึ้นอย่างชัดเจนใน ปี 2567 (สูงสุดในรอบ 7 ปี)

เชื่อมโยงกับ CO₂e Emission
จากผลการวิเคราะห์เชิงเส้นพบว่า:

การสูญเสียป่าปฐมภูมิ 1 ha ส่งผลต่อการปล่อย CO₂e ≈ 568 Mg

ดังนั้น การสูญเสีย 1,000 ไร่ ≈ 160 ha → เพิ่ม CO₂e ≈ 91,000 Mg (หรือ 91,000 ตัน)
